It is observed that above dataset contains 35 columns out of which 4 columns i.e. (Age, Buisness Travel, DailyRate, DistanceFromHome, Marital status) have missing values. 'Attrition' is the target variable ('yes' / 'No' ). Various Datatypes of the columns are object, integer, float.

**Missing value Treatment**

```python
NewNumeric = df[['DistanceFromHome','PercentSalaryHike','TotalWorkingYears','YearsAtCompany','YearsInCurrentRole','YearsSinceLastPromotion','YearsWithCurrManager','TrainingTimesLastYear','NumCompaniesWorked']]
NewNumericMelt = NewNumeric.melt()
plt.figure(figsize=(15,10))
plt.title("Boxplots for Numeric Variables")
bp = sns.boxplot(x = 'variable', y='value', data = NewNumericMelt)
plt.xticks(fontsize = 10, rotation = 'vertical')
plt.show()
```

```python
df.dropna(subset=['Age', 'BusinessTravel', 'DailyRate', 'DistanceFromHome', 'MaritalStatus'], inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
print('Total Number of Unique values in columns:')
for features in df:
    if(features != 'Attrition'):
        print(str(features)+ ': (Datatype:' +str(df[features].dtype) +') : ' +str(len(df[features].unique())))
```

# Exploratory Data Analysis

```python
attrition_count = pd.DataFrame(df['Attrition'].value_counts())
attrition_count
```

```python
plt.pie(attrition_count['Attrition'] , labels = ['No' , 'Yes'] , explode = (0.2,0))
```

From the above pie chart we can see that some employees were attrited but maximum number of employees were not attrited.

```python
plt.figure(figsize = (6,5))
Attrplots = sns.countplot(x = 'Attrition' , data = df)
plt.title('Countplot for Attrition column')
plt.show()
```

From the above observation also we can see the number of attritions. Employees who have resigned their rate is above 100 while the Employees who haven't resigned their rate is above 600.

```python
attrition_dummies = pd.get_dummies(df['Attrition'])
attrition_dummies.head(10)
```

Here I converted Attrition's ['No', 'Yes'] columns into [0,1] i.e. into integer format to perform the following operations.

```python
df = pd.concat([df,attrition_dummies], axis = 1)
df.head()
```

```python
df.drop(['No'], axis = 1, inplace = True)
df.head()
```

```python
#plotting barplot to check gender wise attrition.
plt.figure(figsize = (6,5))
sns.barplot(x = 'Gender', y = 'Yes' , data = df)
plt.title('Gender wise attrition')
plt.show()
```

From the above barplot we can see that males were attrited more than females.

```python
#plotting barplot to check department wise attrition
plt.figure(figsize = (6,5))
sns.barplot(x = 'Department', y ='Yes', data = df)
plt.title('Department wise Attrition')
plt.show()
```

From the above barplot we can see the attrition rate is **maximum in 'Human Resources Department'** after that in'Sales Department' and least in 'Research & Development Department'.

```python
# plotting barplot to check BuisnessTrvale wise attrition
plt.figure(figsize = (6,5))
sns.barplot(x = 'BusinessTravel', y = 'Yes', data = df)
plt.title('BusinessTravel wise Attrition')
plt.show()
```

Attrition rate is maximum for those employees who travels frequently and it is minimum for the employees who don't travel at all.

```python
#checking attrition rate age wise
plt.figure(figsize = (10,8))
plt.xticks(fontsize= 8, rotation='vertical')
sns.barplot(x = 'Age', y = 'Yes', data = df)
plt.title('Age wise Attrition')
plt.show()
```

We can see that attrition rate is maximum in the age group from 18-22 and then in age group 57-58.

```python
#checking attrition rate Total work years wise
plt.figure(figsize = (10,8))
plt.xticks(fontsize= 8, rotation='vertical')
sns.barplot(x = 'TotalWorkingYears', y = 'Yes', data = df)
plt.title('TotalWorkingYears wise Attrition')
plt.show()
```

Employees with 0-3 years of work experience shows maximum rate of attrition and on the other hand aslo Employees with 33-34 years of work experience shows maximum rate of attrition.

```python
plt.figure(figsize = (6,5))
plt.xticks(fontsize= 8, rotation='vertical')
sns.barplot(x = 'EnvironmentSatisfaction', y = 'Yes', data = df)
plt.title('Work Environment wise Attrition')
plt.show()
```

Most employees seem to be satisfied with the working environment.

```python
plt.figure(figsize=(12,12))
sns.heatmap(df.corr(),cbar=True,square=True,fmt='.1f',annot=True,cmap='Reds')
```

**As we can see, there isn't a very strong correlation of the target column with any of the numerical columns. But we can see other correlations such as;**

1) More senior employees have higher total working years (very obvious)
2) Higher performance ratings lead to salary hike percentage to increase
3) The more years an employee puts in, the more their monthly income increases
4) A lot of employees remain in their current role and also under the same manager as years pass by meaning they don't get promotion and this could be a major factor contributing to attrition

**From here, we can deduct that the lack of promotions may be a crucial factor to attritions.**
